```python
# import necessary package
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Task 1 - confirmation analysis

### a) compare the largest and others

```python
def confirmation_compare_largest_others():
    covid_series = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Confirmed_archived_0325.csv')
    covid_series_region =  covid_series[covid_series['Country/Region']=='China'].iloc[:,[0,-1]].sort_values(by="3/23/20", ascending=False)
    largest_pro = covid_series_region.iloc[0,:]
    others = covid_series_region.iloc[1:,:]
    new_df = pd.DataFrame({'Province/State':[largest_pro.iloc[0], 'others'], '3/23/20':[largest_pro[1], others.iloc[:,1].sum()]})
    plt.bar(new_df.iloc[:,0], new_df.iloc[:,1],width=0.45)
    plt.xlabel("Provinces in China")
    plt.ylabel("Number of confrimation cases")
    plt.show()
```

### c) confirmation number distribution by days

```python
def province_confirmation(region='Hubei'):
    covid_series = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Recovered_archived_0325.csv')
    covid_series_region =  covid_series[covid_series['Province/State']==region]
    fig, ax = plt.subplots()
    plt.plot(covid_series_region.iloc[0,4:])
    xticks = ax.xaxis.get_major_ticks()
    for i in range(len(xticks)):
        if i%17!=0:
            xticks[i].set_visible(False) 
            xticks[i]
    xticks[len(xticks)-1].set_visible(True)
    print(region)
    plt.xlabel("Time")
    plt.ylabel("Number of confrimation cases")
    plt.show()
    return covid_series_region
```

## Task 4 - recovery rate vs death rate

### confirmation number

```python
covid_confirmation = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Confirmed_archived_0325.csv')
covid_series_confirmation =  covid_confirmation[covid_confirmation['Country/Region']=='China'].iloc[:,[0,-1]]
covid_series_confirmation.columns = ['Province/State', 'confirmation']
```

### recovery number

```python
covid_recovery = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Recovered_archived_0325.csv')
covid_series_recovery =  covid_recovery[covid_recovery['Country/Region']=='China'].iloc[:,[0,-1]]
covid_series_recovery.columns = ['Province/State', 'recovery']
```

### death number

```python
covid_death = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Deaths_archived_0325.csv')
covid_series_death =  covid_death[covid_death['Country/Region']=='China'].iloc[:,[0,-1]]
covid_series_death.columns = ['Province/State', 'death']
```

### recovery rate

```python
recovery_confirmation = pd.merge(left=covid_series_confirmation, right=covid_series_recovery, on='Province/State')
recovery_confirmation['recovery_rate'] = recovery_confirmation['recovery']/recovery_confirmation['confirmation']
recovery_rate = recovery_confirmation.sort_values(by='recovery_rate', ascending=True)
```

```python
plt.figure(figsize=(50,10))
plt.bar(recovery_rate.iloc[0:,0], recovery_rate.iloc[0:,-1])
plt.xlabel("All provinces except for Hubei in China")
plt.ylabel("The recovery rate")
plt.show()
```

### death rate

```python
death_confirmation = pd.merge(left=covid_series_confirmation, right=covid_series_death, on='Province/State')
death_confirmation['death_rate'] = death_confirmation['death']/death_confirmation['confirmation']
death_rate = death_confirmation.sort_values(by='death_rate', ascending=False)
```

```python
plt.figure(figsize=(50,10))
plt.bar(death_rate.iloc[0:,0], death_rate.iloc[0:,-1])
plt.xlabel("All provinces except for Hubei in China")
plt.ylabel("The death rate")
plt.show()
```

### observation: high death rate is not correlated with low recovery rate.

```python
recovery_death_rate_compare = pd.merge(left=recovery_rate, right=death_rate)
```

### the recovery rate for a specific province over time

```python
def recovery_rate_time(region='Hubei'):
    ## confirmation
    covid_confirmation = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Confirmed_archived_0325.csv')
    covid_series_confirmation =  covid_confirmation[covid_confirmation['Province/State']==region].iloc[:,4:-1]

    ## recovery
    covid_recovery = pd.read_csv('./COVID-19/archived_data/archived_time_series/time_series_19-covid-Recovered_archived_0325.csv')
    covid_series_recovery =  covid_recovery[covid_recovery['Province/State']==region].iloc[:,4:-1]

    ## rate
    rate = covid_series_recovery/covid_series_confirmation

    ## plot
    print(region)
    fig, ax = plt.subplots()
    plt.plot(rate.iloc[0,:])
    xticks = ax.xaxis.get_major_ticks()
    for i in range(len(xticks)):
        if i%17!=0:
            xticks[i].set_visible(False) 
            xticks[i]
    xticks[len(xticks)-1].set_visible(True)
    plt.xlabel("Time")
    plt.ylabel("The recovery rate")
```

```python
recovery_rate_time()
```
